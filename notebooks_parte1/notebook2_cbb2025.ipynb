## Pergunta: Opera√ß√£o Matem√°tica

**Sabendo que o melhor tempo da maratona √© de 2h01m, e a dist√¢ncia da prova √© de 42km, qual a velocidade m√©dia em m/s e km/h do atleta?**

### Resposta:

Para calcular a velocidade m√©dia, precisamos converter o tempo total em segundos e a dist√¢ncia em metros.

- Tempo total: 2 horas e 1 minuto = (2 * 3600) + (1 * 60) segundos
- Dist√¢ncia total: 42 km = 42 * 1000 metros

A f√≥rmula da velocidade m√©dia √© de?



````python
# Convers√£o de tempo e dist√¢ncia
tempo_total_segundos = (2 * 3600) + (1 * 60)  # 2 horas e 1 minuto em segundos
distancia_total_metros = 42 * 1000  # 42 km em metros

# C√°lculo da velocidade m√©dia
velocidade_media_ms = distancia_total_metros / tempo_total_segundos  # em m/s
velocidade_media_kmh = (velocidade_media_ms * 3600) / 1000  # convers√£o para km/h

# Exibindo os resultados

print(f"Velocidade m√©dia em m/s: {velocidade_media_ms}")

print("Velocidade m√©dia em km/h:",  round(velocidade_media_kmh,4))
````
