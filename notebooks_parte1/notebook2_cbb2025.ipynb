## Operadores Relacionais

Operadores relacionais são usados para comparar valores. Eles retornam um valor booleano (`True` ou `False`).

### Igual a (`==`)

Verifica se dois valores são iguais.

```python
a = 5
b = 10
print("a == b:", a == b)  # False
```

### Diferente de (`!=`)

Verifica se dois valores são diferentes.

```python
print("a != b:", a != b)  # True
```

### Maior que (`>`)

Verifica se um valor é maior que o outro.

```python
print("a > b:", a > b)  # False
```

### Menor que (`<`)

Verifica se um valor é menor que o outro.

```python
print("a < b:", a < b)  # True
```

### Maior ou igual a (`>=`)

Verifica se um valor é maior ou igual a outro.

```python
print("a >= b:", a >= b)  # False
```

### Menor ou igual a (`<=`)

Verifica se um valor é menor ou igual a outro.

```python
print("a <= b:", a <= b)  # True
```

### **Praticando:**


````python
a = 10       # Troque os números

b = 11      # Troque os números

print("a == b:", a == b)
print("a != b:", a != b)
print("a < b:", a < b)
print("a > b:", a > b)
print("a <= b:", a <= b)
print("a >= b:", a >= b)
````

## Pergunta: Operação Matemática

**Sabendo que o melhor tempo da maratona é de 2h01m, e a distância da prova é de 42km, qual a velocidade média em m/s e km/h do atleta?**

### Resposta:

Para calcular a velocidade média, precisamos converter o tempo total em segundos e a distância em metros.

- Tempo total: 2 horas e 1 minuto = (2 * 3600) + (1 * 60) segundos
- Distância total: 42 km = 42 * 1000 metros

A fórmula da velocidade média é de?



````python
# Conversão de tempo e distância
tempo_total_segundos = (2 * 3600) + (1 * 60)  # 2 horas e 1 minuto em segundos
distancia_total_metros = 42 * 1000  # 42 km em metros

# Cálculo da velocidade média
velocidade_media_ms = distancia_total_metros / tempo_total_segundos  # em m/s
velocidade_media_kmh = (velocidade_media_ms * 3600) / 1000  # conversão para km/h

# Exibindo os resultados

print(f"Velocidade média em m/s: {velocidade_media_ms}")

print("Velocidade média em km/h:",  round(velocidade_media_kmh,4))
````
