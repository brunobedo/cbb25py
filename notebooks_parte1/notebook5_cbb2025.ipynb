````python
# Criando um DataFrame
data = {
    "Nome": ["Pedro", "Lucas", "Ana", "Maria", "Carlos"],
    "Posi√ß√£o": ["Atacante", "Meio-Campo", "Defensor", "Goleiro", "Atacante"],
    "Gols": [10, 5, 0, 0, 8],
    "Assist√™ncias": [2, 7, 1, 0, 4]
}
df = pd.DataFrame(data)
print(df)
````

## Adicionando uma Nova Coluna

## Ordenando Dados
